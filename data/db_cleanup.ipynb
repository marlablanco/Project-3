```python
import sqlite3
import pandas as pd

from pathlib import Path
```

```python
database_path = "natural_disasters_db.sqlite"
Path(database_path).touch()
```

```python
df = pd.read_csv("DisasterDeclarationsSummaries.csv")
```

```python
df.info()
```

```python
for col in ["declaration_date", "incident_begin_date", "incident_end_date", "disaster_closeout_date", "last_ia_filing_date", "last_refresh"]:
    df[col] = df[col].astype("datetime64").apply(lambda x: x.value)
df
```

```python
conn = sqlite3.connect(database_path)
c = conn.cursor()

c.execute('''CREATE TABLE disasters (
    fema_declaration_string VARCHAR(10),
    disaster_number INT,
    state VARCHAR(2),
    declaration_type VARCHAR(2),
    declaration_date INT,
    fy_declared INT,
    incident_type VARCHAR(32),
    declaration_title VARCHAR(128),
    ih_program_declared INT,
    ia_program_declared INT,
    pa_program_declared INT,
    hm_program_declared INT,
    incident_begin_date INT,
    incident_end_date INT,
    disaster_closeout_date INT,
    tribal_request INT,
    fips_state_code INT,
    fips_county_code INT,
    place_code INT,
    designated_area VARCHAR(64),
    declaration_request_number INT,
    last_ia_filing_date INT,
    last_refresh INT,
    hash VARCHAR(40),
    id VARCHAR(36)
)''')

df.to_sql("disasters", conn, if_exists='append', index=False)

c.close()
conn.close()
```
