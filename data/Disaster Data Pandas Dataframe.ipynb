```python
#Import Dependencies
import sqlite3
import pandas as pd
from pathlib import Path
```

```python
database_path = "natural_disasters_db.sqlite"
```

```python
#Import CSV
disaster_data_load = Path("Resources/1970-2021_DISASTERS.csv")
```

```python
#Read CSV with Pandas
df = pd.read_csv(disaster_data_load)
```

```python
#Display Pandas DataFrame
df.head()
```

```python
#Get rid of unnecessary columns
keep_df = df[["dis_no","year","seq","disaster_group","disaster_subgroup","disaster_type","disaster_subtype","country","iso","latitude","longitude","total_deaths","total_damages"]]
```

```python
keep_df.head()
```

```python

conn = sqlite3.connect(database_path)
c = conn.cursor()
c.execute('DROP TABLE IF EXISTS world_disasters_1970_2021')
c.execute('''CREATE TABLE world_disasters_1970_2021 (
        dis_no VARCHAR(14),
        year INT,
        seq INT,
        disaster_group VARCHAR(30),
        disaster_subgroup VARCHAR(30),
        disaster_type VARCHAR(30),
        disaster_subtype VARCHAR(30),
        country VARCHAR(30),
        iso VARCHAR(3),
        latitude FLOAT,
        longitude FLOAT,
        total_deaths FLOAT,
        total_damages FLOAT
    )''')

keep_df.to_sql("world_disasters_1970_2021", conn, if_exists='append', index=False)

c.close()
conn.close()
        
        
        
```
