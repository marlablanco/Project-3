```python
#Import Dependencies
from flask import Flask, jsonify
import sqlite3
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import func
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
```

```python
# Create an engine for the natural_disasters_db.sqlite database
engine = create_engine("sqlite:///data/natural_disasters_db.sqlite", echo=False)
```

```python
# Reflect Database into ORM classes
Base = declarative_base()

class NaturalDisaster(Base):
    __tablename__ = 'world_disasters_1970_2021'

    dis_no = Column(String, primary_key=True)
    year = Column(Integer)
    seq = Column(Integer)
    disaster_group = Column(String)
    disaster_subgroup = Column(String)
    disaster_type = Column(String)
    disaster_subtype = Column(String)
    country = Column(String)
    iso = Column(String)
    latitude = Column(Float)
    longitude = Column(Float)
    total_deaths = Column(Float)
    total_damages = Column(Float)
```

```python
# Create a session
session = Session(engine)

# Query Database
disasters = session.query(NaturalDisaster).all()

# Print the data for each disaster
for disaster in disasters:
    print(
        f"Disaster Type: {disaster.disaster_type}",
        f"Year: {disaster.year}",
        f"Country: {disaster.country}",
        f"Total Deaths: {disaster.total_deaths}",
    )

# Remember to close the session when you're done with it
session.close()
```
